```python
# import modules
import numpy as np
import pandas as pd
```

Airport geographic is measured in latitude and longitude in the airports csv file. That file will be imported first.

```python
# inputting airport csv file
airports = pd.read_csv('csv_files/airports.csv')
airports.head()
```

## Question 1
We're looking for the northernmost airport in the United States, so let's first review the time zones included in the csv file:

```python
airports.groupby(['tzone']).count()
```

Note that there are airports included in the file that are located in Asia, so we'll want to remove that before moving further

```python
airports2 = airports.query('tzone != "Asia/Chongqing"')
airports2.head()
```

Now we can sort by latitude:

```python
airports2.sort_values(by=['lat'], ascending=False)
```

It looks like the first item (Dillant Hopkins Airport) has its latitude and longitude reversed points reversed (it's located in New Hampshire). The northernmost airport is the **Wiley Post–Will Rogers Memorial Airport** with a latitude of 71.285446.

## Question 2
We're looking for the easternmost airport in the United States.
We'll be looking for a longitude with the highest number:

```python
#reusing previously created object
airports2.sort_values(by=['lon'], ascending=False)
```

Looking at the data, Eareckson Air Station is actually the westernmost airport as it belongs to a different time zone. Also, because the Dillant information is backwards, we will ignore that as well. So the easternmost airport is the **Eastport Municipal Airport** with a longitude of -67.012694.

## Question 3
On February 12th, 2013, which New York area airport had the windiest weather?
For this question, we'll need to import the weather csv file:

```python
# inputting weather csv file
weather = pd.read_csv('csv_files/weather.csv')
weather.head()
```

To make sure we're looking at the right date, we'll want to filter for the year, month and date:

```python
weather2 = weather[(weather.year == 2013) & (weather.month == 2) & (weather.day == 12)].sort_values(by=['wind_speed'], ascending=False)
weather2
```

There seems to be an outlier (likely inaccurate) data point at a wind speed of 1048.36058, so we'll have to remove it:

```python
weather2 = weather2[(weather2.wind_speed < 1000)].sort_values(by=['wind_speed'], ascending=False)
weather2
```

Because there are multiple wind speed recordings for each airport on that day, we'll have to calculate the average wind speed for each airport:

```python
weather3 = weather2.groupby(['origin']).mean().sort_values(by=['wind_speed'], ascending=False)
weather3
```

We can see that with an average of 14.960140, **La Guardia Airport (LGA)** had the highest average wind speed for February 12, 2013.
